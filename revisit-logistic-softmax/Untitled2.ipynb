```python
import matplotlib.pyplot as plt

# Example data
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
values = [5, 7, 3, 8, 6]

# Create a bar graph
plt.figure(figsize=(10, 6))  # Optional: to set the figure size
plt.bar(names, values, color='blue')

# Add titles and labels
plt.title('Bar Graph with Names as X-Axis')
plt.xlabel('Names')
plt.ylabel('Values')

# Show the plot
plt.show()
```
