```python
meta_train_data, meta_valid_data, meta_test_data = provide_data(dataset='berkeley', data_dir="./filelists/sensor_data/")

print("type meta_train_data :",type(meta_train_data))
print("len meta_train_data :", len(meta_train_data))
print("all meta_train_data[i] are of len 2 :", all([len(meta_train_task)==2 for meta_train_task in meta_train_data]))

print(meta_train_data[0][0].shape)
print(meta_train_data[0][1].shape)
print(meta_train_data[35][0].shape)
print(meta_train_data[35][1].shape)

print()

print("type meta_valid_data :",type(meta_valid_data))
print("len meta_valid_data :", len(meta_valid_data))
print("all meta_valid_data[i] are of len 2 :", all([len(meta_valid_task)==4 for meta_valid_task in meta_valid_data]))

print(meta_valid_data[0][0].shape)
print(meta_valid_data[0][1].shape)
print(meta_valid_data[0][2].shape)
print(meta_valid_data[0][3].shape)
print(meta_valid_data[9][0].shape)
print(meta_valid_data[9][1].shape)
print(meta_valid_data[9][2].shape)
print(meta_valid_data[9][3].shape)

print()

print("type meta_test_data :",type(meta_test_data))
print("len meta_test_data :", len(meta_test_data))
print("all meta_test_data[i] are of len 2 :", all([len(meta_test_task)==4 for meta_test_task in meta_test_data]))

print(meta_test_data[0][0].shape)
print(meta_test_data[0][1].shape)
print(meta_test_data[9][0].shape)
print(meta_test_data[9][1].shape)

print()

print("Meta valid data == meta test data :", meta_valid_data == meta_test_data)

print()
print(meta_train_data[0][0][0])
print(meta_test_data[0][0][0])
```

```python
def convert_big_tuple(meta_data):
    arrX_list = []
    arrY_list = []
    for (arrX, arrY) in meta_data:
        arrX_list.append(arrX)  # (30, 11)
        arrY_list.append(arrY)  # (30,)

    big_X_np = np.stack(arrX_list, axis=0)
    big_Y_np = np.stack(arrY_list, axis=0)

    big_X_torch = torch.from_numpy(big_X_np)
    big_Y_torch = torch.from_numpy(big_Y_np)

    return big_X_torch, big_Y_torch

meta_X, meta_label = convert_big_tuple(meta_train_data)
print(meta_X.shape, meta_label.shape)
```
