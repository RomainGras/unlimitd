```python
def toy_function(x, y, A=1, B=1, C=1, noise_std=0.1):
    A_random = A + np.random.normal(0, 0.2)
    B_random = B + np.random.normal(0, 0.2)
    C_random = C + np.random.normal(0, 0.1)
    
    result = A_random * x**2 + B_random * y**2 + C_random * np.sin(x * y)
    
    # Add Gaussian noise to the result to simulate measurement noise
    result += np.random.normal(0, noise_std)
    return result
```

```python
import matplotlib.pyplot as plt

# Create a meshgrid for inputs x, y
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Calculate function values on the grid
Z = toy_function(X, Y)

# Plotting the function surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='inferno')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Toy Nonlinear Coupled Function')

plt.show()

```
