```python
def load_datapoints_shapenet(file_path):
    x, y = pickle.load(open(file_path, 'rb'))

    x = np.array(x)
    x = x[..., np.newaxis]

    y = np.array(y)
    y = y[:, :, -1, None]
    y = y*2*np.pi
    y = np.concatenate([np.cos(y), np.sin(y), y], axis=-1)
    
    return x, y
```
