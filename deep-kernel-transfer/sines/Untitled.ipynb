```python
import matplotlib.pyplot as plt

# Example data
points_a = [(1, 2), (2, 3), (3, 5)]
points_b = [(2, 5), (3, 6), (4, 8)]

# Unpack points into separate lists for plotting
x1, y1 = zip(*points_a)  # Unpack series a
x2, y2 = zip(*points_b)  # Unpack series b

# Create scatter plots for both series
plt.scatter(x1, y1, color='blue', marker='o', label='Series A')
plt.scatter(x2, y2, color='red', marker='x', label='Series B')

# Draw lines between corresponding points
for (xa, ya), (xb, yb) in zip(points_a, points_b):
    plt.plot([xa, xb], [ya, yb], 'gray', linestyle='--')  # Connecting lines

# Add labels and a legend
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Mapping Points from Series A to B')
plt.legend()

# Show the plot
plt.show()
```
